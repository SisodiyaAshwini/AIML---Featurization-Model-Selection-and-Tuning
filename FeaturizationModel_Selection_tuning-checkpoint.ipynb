```python
# Handling Outliers: Let's replace the outliers with zscore. 
# Here removing the dependent column i.e. strength.

data = dataset.loc[:,'cement':'age']

# Applying zscore to remove outliers
from scipy.stats import zscore

data_z = dataset.apply(zscore)
```

```python
def Remove_outliers(x):
    Q3 = x.quantile(0.75)
    Q1 = x.quantile(0.25)
    low_lim = Q1 - (Q3 - Q1)
    up_lim =  Q3 + (Q3 - Q1)
    return list( map( lambda x: x if(x < up_lim and x > low_lim) else np.nan, x.values) )

data_z["slag"] = Remove_outliers(data_z["slag"])
data_z["age"] = Remove_outliers(data_z["age"])
data_z["water"] = Remove_outliers(data_z["water"])
data_z.fillna(dataset.mean(), inplace=True)
sns.pairplot(data_z, diag_kind='kde')
data_z.boxplot()
```

```python
# Q2)  a. Identify opportunities (if any) to create a composite feature, drop a feature etc. 
# b. Decide on complexity of the model, should it be simple linear model in terms of parameters or would a quadratic or higher degree help
# c. Explore for gaussians. If data is likely to be a mix of gaussians, explore individual clusters and present your findings in terms of the independent attributes and their suitability to predict strength (5 marks)

#Explanation

# Divide data into traning and test set.
x = data_z.iloc[:,0:8]
y = data_z.iloc[:,8]

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test, = train_test_split(x,y, test_size = 0.3, random_state = 10)
```

```python
from sklearn import svm
clr = svm.SVR()
clr.fit(x_train, y_train)
y_pred = clr.predict(x_test)
```

```python
# As we are using regression model, we can use joinplot to see what we have achieved. We are doing correlation analysis. 
import seaborn as sns
from scipy.stats import pearsonr

sns.set(style="darkgrid", color_codes=True)
with sns.axes_style("white"):
    sns.jointplot(x=y_test, y=y_pred, kind="reg", color="k", stat_func=pearsonr)
```

```python
# PLot shows that accuarcy is 90%, which is quite good. Let's try to improve this by dropping column(s).
# In pairplot, slag doesn't have strong relationship with strength, so lets drop slag columns and create joinplot.

data_2 = dataset.drop("superplastic", axis = 1)
data_2_z = data_2.apply(zscore)

data_2_z["age"] = Remove_outliers(data_2_z["age"])
data_2_z["water"] = Remove_outliers(data_2_z["water"])
data_2_z["slag"] = Remove_outliers(data_2_z["slag"])
data_2_z.fillna(dataset.mean(), inplace=True)

x = data_2_z.iloc[:,0:7]
y = data_2_z.iloc[:,7]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 10)

clr.fit(x_train, y_train)

y_pred = clr.predict(x_test) 

sns.set(style="darkgrid", color_codes=True)
with sns.axes_style("white"):
    sns.jointplot(x=y_test, y=y_pred, kind="reg", color="k", stat_func=pearsonr)
```

```python
# Dropping a column does not impact the accuracy. Hence proves the weak correlation

# Achieved 90% on test data. The low score is may be due to gussians. 
# Let's see how many clusters are present by using Kmeans clustering. From pairplot we can visualize 3 to 15 clusters can be present. 
# For this we are restricting clusters between 2 to 6.

from sklearn.cluster import KMeans

cluster_range = range(2,6)
cluster_errors = []
for num_clusters in cluster_range:
    clusters = KMeans(num_clusters, n_init=5)
    clusters.fit(dataset)
    labels = clusters.labels_
    centroids = clusters.cluster_centers_
    cluster_errors.append(clusters.inertia_)
cluster_df = pd.DataFrame({"num_clusters":cluster_range, "cluster_errors":cluster_errors})
cluster_df[0:8]
```

```python
#Elbow plot : to choose number of clusters

plt.figure(figsize=(12,6))
plt.plot(cluster_df.num_clusters, cluster_df.cluster_errors, marker="o")
```

```python
# From elbow plot it clearly came out as the number of clusters we would be exploring is either 3 or 4. 
# Let us create 3 cluster analysis

data_3_z = dataset.apply(zscore)

data_3_z["age"] = Remove_outliers(data_3_z["age"])
data_3_z["water"] = Remove_outliers(data_3_z["water"])
data_3_z.fillna(dataset.mean(), inplace=True)

cluster = KMeans(n_clusters = 4, random_state=2345)
cluster.fit(data_3_z)

prediction = cluster.predict(data_3_z)

data_3_z["Group"] = prediction

data_3_z_copy = data_3_z.copy(deep = True)
```

```python
# Q5) Tuning the model
# a. Algorithms that you think will be suitable for this project  
# b. Techniques employed to squeeze that extra performance out of the model without making it overfit or underfit
# c. Model performance range at 95% confidence level

# Explanation:

# Algorithm which will sutiable for this project is SVR

# Lets try tuneup Hyperparamter
# Divide data into three parts Train, Validate and test.

x = data_2_z.iloc[:,0:7]
y = data_2_z.iloc[:,7]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 10)

# Grid Search

from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.decomposition import PCA

pipe_svc = Pipeline([('scl', StandardScaler()), ('pca', PCA()), ('svc', SVC())])

param_grid = {'pca__n_components':[6,7],'svc__C':[0.001, 0.01, 0.1, 1, 10, 100], 'svc__gamma': [0.001,0.01, 0.1, 1, 10], 'svc__kernel':['rbf','poly']}

grid = GridSearchCV(pipe_svc, param_grid = param_grid, cv=5)

lab_enc = preprocessing.LabelEncoder()
#x_encoded = lab_enc.fit_transform(x_train)
y_encoded = lab_enc.fit_transform(y_train)

grid.fit(x_train, y_encoded)

print(" Best cross_validation accuracy: {:.2f}". format(grid.best_score_))
print(" Best parameters: ", grid.best_params_)
print(" Test set accuaracy: {:.2f}". format(grid.score(x_test, y_test)))
```
